```python
import gzip

def build_synonym_dict(gene_info_gz_path):
    synonym_dict = {}
    with gzip.open(gene_info_gz_path, 'rt') as f:
        for line in f:
            if line.startswith("#"):
                continue  # Skip header lines
            # Typically, columns in gene_info are tab-delimited.
            cols = line.strip().split('\t')
            
            # For NCBI gene_info, columns might be:
            # tax_id, GeneID, Symbol, LocusTag, Synonyms, dbXrefs, chromosome, map_location, ...
            tax_id, gene_id, symbol, locus_tag, synonyms_str = cols[:5]
            
            # We only want human genes, for example:
            if tax_id != '9606':
                continue
            
            synonyms = synonyms_str.split('|')
            
            # Add the standard symbol as well as known synonyms to the dictionary
            all_keys = set([symbol] + synonyms)
            for key in all_keys:
                if key:  # skip empty strings
                    synonym_dict[key.upper()] = symbol  # uppercase to allow case-insensitive matching
    return synonym_dict
```
